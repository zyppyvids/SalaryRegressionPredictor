# Purpose of this model
This model uses data from a quiz done in the Bulgarian IT community. The purpose is to be able to predict salary based on years of experience in the current tech stack and years of experience overall.

<b>To use just write 'regression.predict([[\<starting_salary\>, \<current_experience\>, \<overall_experience\>, \<people_you_manage\>, {1/0 for the position you are working on}]])'</b>

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
salaries = pd.read_csv("data.csv")
salaries
```

```python
salaries_copy = salaries.copy()

# We filter the faulty rows (current experience cannot be higher than overall experience, salary cannot be 0)
salaries_copy = salaries_copy[(salaries_copy.curexp <= salaries_copy.exp) & (salaries_copy.salary != 0) & (salaries_copy.stsalary != 0) & (salaries_copy.stsalary <= salaries_copy.salary)]
salaries_copy
```

```python
# We fill NaN column values
salary_mean = salaries_copy.salary.mean()
curexp_mean = salaries_copy.curexp.mean()
exp_mean = salaries_copy.exp.mean()

salaries_copy.salary = salaries_copy.salary.fillna(salary_mean)
salaries_copy.curexp = salaries_copy.curexp.fillna(curexp_mean)
salaries_copy.exp = salaries_copy.exp.fillna(exp_mean)

salaries_copy = salaries_copy[(salaries_copy.curexp <= salaries_copy.exp)]
salaries_copy
```

```python
salary_mean
```

```python
curexp_mean
```

```python
exp_mean
```

```python
# IQR Outliers removal happens here
# SALARY
Q1 = np.percentile(salaries_copy.salary, 25,
                   method = 'midpoint')
 
Q3 = np.percentile(salaries_copy.salary, 75,
                   method = 'midpoint')
IQR = Q3 - Q1

upper_boundary = Q3 + 1.5 * IQR
lower_boundary = Q1 - 1.5 * IQR

salaries_copy = salaries_copy[(salaries_copy.salary <= upper_boundary) & (salaries_copy.salary >= lower_boundary)]

# EXPERIENCE
Q1 = np.percentile(salaries_copy.exp, 25,
                   method = 'midpoint')
 
Q3 = np.percentile(salaries_copy.exp, 75,
                   method = 'midpoint')
IQR = Q3 - Q1

upper_boundary = Q3 + 1.5 * IQR
lower_boundary = Q1 - 1.5 * IQR

salaries_copy = salaries_copy[(salaries_copy.exp <= upper_boundary) & (salaries_copy.exp >= lower_boundary)]


# CURRENT EXPERIENCE
Q1 = np.percentile(salaries_copy.curexp, 25,
                   method = 'midpoint')
 
Q3 = np.percentile(salaries_copy.curexp, 75,
                   method = 'midpoint')
IQR = Q3 - Q1

upper_boundary = Q3 + 1.5 * IQR
lower_boundary = Q1 - 1.5 * IQR

salaries_copy = salaries_copy[(salaries_copy.curexp <= upper_boundary) & (salaries_copy.curexp >= lower_boundary)]

# PEOPLE MANAGE
Q1 = np.percentile(salaries_copy.peoplemanage, 25,
                   method = 'midpoint')
 
Q3 = np.percentile(salaries_copy.peoplemanage, 75,
                   method = 'midpoint')
IQR = Q3 - Q1

upper_boundary = Q3 + 1.5 * IQR
lower_boundary = Q1 - 1.5 * IQR

salaries_copy = salaries_copy[(salaries_copy.peoplemanage <= upper_boundary) & (salaries_copy.peoplemanage >= lower_boundary)]
salaries_copy
```

```python
# Label encoding happens here
salaries_dummies = pd.get_dummies(salaries_copy.position)

salaries_merged = pd.concat([salaries_copy, salaries_dummies], axis = 1)

salaries_filtered = salaries_merged.drop(['position', 'QA'], axis = 1)
salaries_filtered
```

```python
# Plotting happens here
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Experience')
plt.ylabel('Salary')

plt.scatter(salaries_filtered.exp, salaries_filtered.salary, marker = '+')
```

```python
plt.xlabel('Current Experience')
plt.ylabel('Salary')

plt.scatter(salaries_filtered.curexp, salaries_filtered.salary, marker = '+')
```

```python
# Data splitting happens here
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(salaries_filtered[['stsalary', 'curexp', 'exp', 'peoplemanage', 'DevOps', 'Developer', 'Hardware', 'Support', 'Team Lead']].values, salaries_filtered.salary.values, test_size = 0.2)
```

```python
# The model creation happens here
regression = linear_model.LinearRegression()
regression.fit(X_train, y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
regression.predict([[2000, 4, 4, 1, 0, 1, 0, 0, 0]])
```

```python
regression.score(X_test, y_test)
```

```python
# Exporting happens here
import pickle
with open('salaries_model', 'wb') as file:
    pickle.dump(regression, file)
```
